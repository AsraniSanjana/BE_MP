```python
import PyPDF2

# Replace 'your_file.pdf' with the name of your uploaded PDF file
pdf_file_name = 'chachu_03Sep2023.pdf'

# Open the PDF file
pdf_file = open(pdf_file_name, 'rb')

# Create a PDF reader object using PdfReader
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Initialize a variable to store extracted text
extracted_text = ''

# Loop through each page in the PDF
for page in pdf_reader.pages:
    # Extract text from the current page
    extracted_text += page.extract_text()

# Close the PDF file
pdf_file.close()

# Print the extracted text
print(extracted_text)
```

```python
!pip install pdfplumber
!pip install parse

import re

import parse
import pdfplumber
import pandas as pd
from collections import namedtuple

Line = namedtuple('Line', 'company_id company_name doctype reference currency voucher inv_date due_date open_amt_tc open_amt_bc current month1 month2 month3')

company_re = re.compile(r'(V\d+) (.*) Phone:')
line_re = re.compile(r'\d{2}/\d{2}/\d{4} \d{2}/\d{2}/\d{4}')
file = 'Sample Report Pythonic.pdf'

lines = []
total_check = 0

with pdfplumber.open(file) as pdf:
    pages = pdf.pages
    for page in pdf.pages:
        text = page.extract_text()
        for line in text.split('\n'):
            print(line)
            comp = company_re.search(line)
            if comp:
                vend_no, vend_name = comp.group(1), comp.group(2)

            elif line.startswith('INVOICES'):
                doctype = 'INVOICE'

            elif line.startswith('CREDITNOTES'):
                doctype = 'CREDITNOTE'

            elif line_re.search(line):
                items = line.split()
                lines.append(Line(vend_no, vend_name, doctype, *items))

            elif line.startswith('Supplier total'):
                tot = float(line.split()[2].replace(',', ''))
                total_check += tot


df = pd.DataFrame(lines)
df.head()


df.info()

df['inv_date'] = pd.to_datetime(df['inv_date'])
df['due_date'] = pd.to_datetime(df['due_date'])
for col in df.columns[-6:]:
    df[col] = df[col].map(lambda x: float(str(x).replace(',', '')))


df['open_amt_bc'].sum()
total_check
df.to_csv('invoices.csv', index=False)
```
